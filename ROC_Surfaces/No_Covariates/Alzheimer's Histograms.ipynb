```python
# C:\Mark\uni\Uni Work\5th Year\Diss\ROC Surfaces\adni_noNA.csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from statsmodels.distributions.empirical_distribution import ECDF
import matplotlib as mpl
import seaborn as sns

%matplotlib notebook
```

```python
df = pd.read_csv('C:\\Mark\\uni\\Uni Work\\5th Year\\Diss\\ROC Surfaces\\adni_noNA.csv')
df
```

```python
d1 = df[df['DX'] == 1]
d2 = df[df['DX'] == 2]
d3 = df[df['DX'] == 3]
```

```python
fig, ax = plt.subplots(3,4)
#mpl.rcParams['text.usetex'] = True

#lafont = {'fontstyle':'Computer Modern'}

seaborneplot = True

counti = 0
countj = 0
for i in [d1, d2, d3]:
    for j in ['tau', 'abeta', 'Hipp', 'HCI']:
        
        if seaborneplot:
            sns.histplot(data=i, x=j, kde=True, bins=20, ax=ax[counti, countj])
        else:
            ax[counti, countj].hist(i[j], bins=20, density=False)
        
        
        #ax[counti, countj].set_title(f'd{counti+1} {j}')
        ax[counti, countj].tick_params(left = False, right = False , labelleft = False, labelbottom=True)
        ax[counti, countj].spines['right'].set_visible(False)
        ax[counti, countj].spines['top'].set_visible(False)
        ax[counti, countj].spines['left'].set_visible(False)
        
        if countj == 0 and counti == 0:
            ax[counti, countj].set_title('$\mathrm{Tau}$')
        elif countj == 1 and counti == 0:
            ax[counti, countj].set_title('Abeta')
        elif countj == 2 and counti == 0:
            ax[counti, countj].set_title(j)
        elif countj == 3 and counti == 0:
            ax[counti, countj].set_title(j)
            
        if countj == 0 and counti == 0:
            ax[counti, countj].set_ylabel('D1', rotation=0, ha='right')
        elif countj == 0 and counti == 1:
            ax[counti, countj].set_ylabel('D2', rotation=0, ha='right')
        elif countj == 0 and counti == 2:
            ax[counti, countj].set_ylabel('D3', rotation=0, ha='right')
        else:
            ax[counti, countj].set_ylabel(None)
        
        if countj == 0:
            ax[counti, countj].set_xlim(0,2000)
            ax[counti, countj].set_ylim(0,50)
        elif countj == 1:
            ax[counti, countj].set_xlim(0,2000)
            ax[counti, countj].set_ylim(0,60)
        elif countj == 2:
            ax[counti, countj].set_xlim(3,12)
            ax[counti, countj].set_ylim(0,80)
        elif countj == 3:
            ax[counti, countj].set_xlim(0,45)
            ax[counti, countj].set_ylim(0,90)
        
        countj+=1
        
    countj=0
    counti+=1
```
