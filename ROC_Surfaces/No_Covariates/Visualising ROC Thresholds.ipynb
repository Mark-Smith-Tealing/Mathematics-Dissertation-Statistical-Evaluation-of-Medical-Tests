# Threshold Visualisation for ROC Surfaces

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math

def norm(mu, var, statsnorm=False):
    """
    gives the coordinates required to plot a normal distribution given the parameters provided.
    
    Input:
    >>> mu  -> the mean of the distribution
    >>> var -> the variance of the distribution
    >>> statsnorm -> to caluclate using stats.norm.pdf or not
    
    Output:
    >>> x  -> the x values of the distribution (within 4 s.ds of the mean)
    >>> y2 -> the y values calculated explicitly
    >>> y  -> the y values calculated using stats.norm
    """
    
    sigma = math.sqrt(var)
    x = np.linspace(mu-3.5*sigma, mu +3.5 *sigma, 300)
    
    if statsnorm:
        y = stats.norm.pdf(x, mu, sigma)
    else:
        y = [np.exp(-1/2*((i-mu)/sigma)**2)/ (sigma*np.sqrt(2 * np.pi)) for i in x]
    return x, y

mu1, mu2, mu3 = -2, 5, 13
var1, var2, var3 = 6, 6, 6

norm1x, norm1y = norm(mu1, var1)
norm2x, norm2y = norm(mu2, var2)
norm3x, norm3y = norm(mu3, var3)

fig, ax = plt.subplots(1,1)

ax.plot(norm1x, norm1y, 'blue')
ax.plot(norm2x, norm2y, 'black')
ax.plot(norm3x, norm3y, 'green')
ax.plot(np.zeros(10),np.linspace(-2,1,10), 'black', linestyle='dashed')
ax.plot(9*np.ones(10),np.linspace(-2,1,10), 'black', linestyle='dashed')
ax.tick_params(left = False, right = False , labelleft = False, labelbottom=False)
ax.set_ylim(0,0.17)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.show()
# ax.plot(normhx, normhy, 'r-')
# ax.plot(normdx, normdy, 'r-')
```
